```python
last = names.name.str.get(-1)
```

```python
names["last"] = last
```

```python
cts = names.groupby(['year','last'])['count'].sum().unstack().fillna(0)
```

```python
cts["a y n l d".split()].div(cts.sum(axis=1), axis=0).plot()
```
